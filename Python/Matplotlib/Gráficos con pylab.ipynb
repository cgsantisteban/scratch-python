### Gráficos con pylab (matplotlib)

#### Dibujamos el punto (4,5) de color verde

```python
import pylab

#Creamos el gráfico de puntos
pylab.scatter(4,5, c='g')

#Mostramos el gráfico
pylab.show()
```

#### Dibujamos los puntos (-4,10), (9,1) y (5,7) de color magenta y añadimos un título.

```python
import pylab


#Definimos las coordenadas X de los puntos
coordenadas_x = [-4,9,5]

#Definimos las coordenadas Y de los puntos
coordenadas_y = [10,1, 7]

#Dibujamos los puntos
pylab.scatter(coordenadas_x, coordenadas_y, c='m')

#Ponemos un título al gráfico
pylab.title("Gráfico de puntos")

#Ponemos una etiqueta al eje X
pylab.xlabel("X")

#Ponemos una etiqueta al eje Y
pylab.ylabel("Y")

#Mostramos el gráfico
pylab.show()
```

#### Dibujamos un rectángulo de lados 8 y 14 en las coordenadas (2,3) con color de relleno amarillo

```python
import pylab

#Definimos los límites del eje X de -15 a 15
pylab.xlim(0,14)

#Definimos los límites del eje Y de -15 a 15
pylab.ylim(0,20)


# Dibujamos un rectángulo de lados 8 y 14 en las coordenadas (2,3) con color de relleno amarillo
rectangulo = pylab.Rectangle((2,3), 8, 14, fc='y')

#Añadidmos el círculo al gráfico
pylab.gca().add_patch(rectangulo)


#Mostramos el gráfico
pylab.show()
```

#### Dibujamos un círculo de radio 10, centrado las coordenadas (4,4) y con color de relleno rojo

```python
import pylab

#Fijamos la anchura y altura de la imagen del gráfico. Esto es opcional
anchura = 6
altura = 6
pylab.figure(figsize=(anchura, altura))


#Definimos los límites del eje X de -15 a 15
pylab.xlim(-10,15)

#Definimos los límites del eje Y de -15 a 15
pylab.ylim(-10,15)

#Mostramos subdivisiones en los ejes
pylab.grid(True)

#Definimos un círculo de radio 4 y centrado en las coordenadas (5,7) de color amarillo
circulo = pylab.Circle((4, 4), radius=10, fc='r')

#Añadidmos el círculo al gráfico
pylab.gca().add_patch(circulo)

#Mostramos el gráfico
pylab.show()
```

#### Dibujamos una línea poligonal con los puntos (2,4), (7,10) y (6,-5)

```python
import pylab


#Definimos los límites del eje X de 0 a 6
pylab.xlim(0,10)

#Definimos los límites del eje Y de 0 a 6
pylab.ylim(-10,12)

#Definimos los vértices
vertices = [[2, 4], [7, 10], [6, -5]]

#Creamos la línea poligonal
linea = pylab.Polygon(vertices, closed=None, fill=None)

#Añadimos la línea al gráfico
pylab.gca().add_patch(linea)

#Mostramos el gráfico
pylab.show()
```

#### Dibujamos un polígono cerrado de vértices (1,1), (4,5) y (4,1) y relleno de color magenta

```python
import pylab


#Título del gráfico
pylab.title("Polígono") 

#Título del eje X
pylab.xlabel("x")

#Titulo del eje Y
pylab.ylabel("y")

#Definimos los límites del eje X de 0 a 6
pylab.xlim(0,6)

#Definimos los límites del eje Y de 0 a 6
pylab.ylim(0,6)

#Mostramos subdivisiones en los ejes
pylab.grid(True)

#Definismo los vértices
vertices = [[1, 1], [4, 5], [4, 1]]

#Creamos el polígono
poligono = pylab.Polygon(vertices, closed=True, fill=True, fc='m')

#Añadimos el polígono al gráfico
pylab.gca().add_patch(poligono)

#Mostramos el gráfico
pylab.show()
```
